```python
import google.generativeai as genai 
import PIL
from PIL import Image
import os 
```

```python
defect_image = Image.open('OIP.jfif')
```

```python
defect_image
```

```python
gemini_api_key = os.getenv('Google_API_Key2')
genai.configure(api_key=gemini_api_key)
model = genai.GenerativeModel('gemini-2.5-flash-lite')
```

```python
prompt ='Identify the structural defect in the image and suggest remedies to fix this.'
```

```python
model.generate_content([prompt,defect_image],generation_config={'temperature':0.7})
```
